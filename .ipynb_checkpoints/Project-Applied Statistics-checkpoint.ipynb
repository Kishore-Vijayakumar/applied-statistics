## 2. Question: An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

##### A. Probability that none of the items are defective?


```python
#The question satisfies Binomial Distribution, therefore we use python package scipy.stats
import numpy as np
import scipy.stats as stats
```

```python
p1 = 0.05 #Probability of failed item manufactured
n1 = 10   #Number of samples
k1 = np.arange(0,11) #Sample array of possible defective items
print(k1) 
binom1 = stats.binom.pmf(k1,n1,p1) #Using the probability mass function we can obtain the probability of each outcome and store it in an array
binom1
```

```python
#Solution 1

#Probability for none of the items are defective will be Probability of 0 defective item, therefore binom1[0]
print(' Probability that none of the items are defective = %1.4f'% binom1[0])
```

##### B. Probability that exactly one of the items is defective?

```python
#Solution 2
#Probability for exactly one of the items is defective will be Probability of 1 defective item, therefore binom1[1]
print('Probability that exactly one of the items is defective = %1.4f' % binom1[1])
```

##### C. Probability that two or fewer of the items are defective?


```python
#Solution 3
#Probability for two or fewer of the items are defective will be Sum of probabilities of 0,1,2 defective items,
#therefore binom1[0]+binom1[1]+binom1[2]
sum_fewer_or_two = binom1[0]+binom1[1]+binom1[2]
print('Probability that two or fewer of the items are defective = %1.4f' %sum_fewer_or_two)
```

##### D. Probability that three or more of the items are defective ?

```python
#Solution 4
#Probabilities for three or more of the items are defective will be Sum of probabilities of 3,4,5,6,7,8,9,10 defective items,
#We can find it by either adding the probabilities from 3 to 10 defective items or we can subtract the probability of 0,1,2 defective items from the total probability of 1
# i.e 1-sum of probability of (0,1,2) defective items

sum_more_or_two = 1 - sum_fewer_or_two
print('Probability that three or more of the items are defective = %1.4f'%sum_more_or_two)
```

## 3. Question: A car salesman sells on an average 3 cars per week

##### A. Probability that in a given week he will sell some cars.

```python
#The question satisfies Poisson Distribution, therefore we use python package scipy.stats

rate1 = 3 #average rate, denoted with lambda in the formula for poisson distribution

k2 = np.arange(0,11) #a numpy array of different no. of successes ranging from 0 to 19 to construct a probability distribution

k2
```

```python
#calculating the distribution and storing the distribution of probablitites in an array

poisson1 = stats.poisson.pmf(k2,rate1)
poisson1
```

```python
#Solution 1

#The probability that in a given week he will sell some cars means he will not sell Zero cars, 
#so we can find the probability of some cars by calculating probability of selling zero cars and subtracting it from 1 (1 - probability[0]) 

some_cars = 1 - poisson1[0]
print('Probability that in a given week he will sell some cars = %1.4f'%some_cars)
```

##### B. Probability that in a given week he will sell 2 or more but less than 5 cars

```python
#Solution 2
#We can calculate probability that in a given week he will sell 2 or more but less than 5 cars, by subtracting the sum of probability of selling 2,3 and 4 from total probability 1

sum_more2_less5 = 1 - (poisson1[2]+poisson1[3]+poisson1[4])
print('Probability that in a given week he will sell 2 or more but less than 5 cars = %1.4f'%sum_more2_less5)
```

##### C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per-week

```python
#Solution 3
#To find cumilative probability we can use  stats.poisson.cdf
poisson1_cumilative = stats.poisson.cdf(k2,rate1)
print(poisson1_cumilative) #The cumilative probability of cars sold per week
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(k2,poisson1_cumilative,'-o')
plt.title('Poisson: $\lambda$ = %i ' % rate1)
plt.xlabel("Number of Cars sold per week")
plt.ylabel("Cumulative Probability of Cars sold per-week")
plt.show()
```

Question: Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed, marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Suppose that you place order with the bot and two friends of yours independently place orders with the same bot. Answer the following questions.

```python
#Solution 2
print('Probability that none of the three orders will be recognised correctly = %1.4f'%binom2[0])
```

## 5. Question: A group of 300 professionals sat for a competitive exam. The results show the information of marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a normal distribution. Answer the following questions.

##### A. What is the percentage of students who score more than 80

```python
mean1 = 60 
sd1 = 12
```

```python
#Solution 1
#We can solve normal distribution by standardizing the values and finding the cdf
z1 = (80-mean1)/sd1 #Standardizing
print('Z-Value = %1.4f'%z1)
#We can use Z value to calculate the probability or the area under curve to the left of this z value
normal1 = 1 - stats.norm.cdf(z1) #The probability (area) to the right is calculated as (1 - probability to the left)

print('Percentage of students who score more than 80 = %1.2f%%' %(normal1*100))
```

##### B. What is the percentage of students who score less than 50.

```python
#Solution 2
z2 = (50-mean1)/sd1 
print('Z-Value = %1.4f'%z2)
#We can use Z value to calculate the probability or the area under curve to the left of this z value
normal2 = stats.norm.cdf(z2)
print('Percentage of students who score less than 50 = %1.2f%%' %(normal2*100))
```

##### C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?

```python
#Solution 3
normal3 = 100-10 #The percentage of students that are not awarded distinction
normal3=normal3/100
#Since we got the percentage, we can calculate the z value using the stats.norm.ppf (Percent Point Function) function
z3 = stats.norm.ppf(normal3)
print('The Z value = %1.4f'%z3)
#Since we got Z value we can find the x or the corresponding mark from the equation z = (x-mean)/standard_deviation
dist = (z3*sd1)+mean1
dist
print('The distinction mark if the highest 10%% of students are to be awarded distinction = %2.2f'%dist)
```

## 6. Question: Explain 1 real life industry scenario [other than the ones mentioned above] where you can use the concepts learnt in this module of Applied statistics to get a data driven business solution.

Company-X is an EV Charging station providers. They have huge network of charging stations in the state of Kerala. Since the demand for electric vehicles has been growing, the company has decided to increase the number of charging slots at charging stations that have a high demand. They drafted a criteria for this upgrading, which had a condition that the probability of number of cars arriving at a charging station in an hour must be greater than the probability of cars the station can hold in an hour.

We can use poisson distribution to find the probability. 
For eg:

Question 1) On an average 3 cars arrive every 1 hour at a charging station which have a capacity of 4 cars.
What will be the probability that more than 4 cars will arrive at the station in a given hour?


```python
#solution 1
rate2 = 3 #Average no: of cars in an hour
k4 = np.arange(0,11)
k4
```

```python
poisson2 = stats.poisson.pmf(k4,rate2)
poisson2
```

```python
#We need to find probability of more than 4 cars, we need to find the sum of probability of 5,6,7,8,9,10 cars arriving,
#for that we can use the cumulative distribution function.
poisson2_cumulative = stats.poisson.cdf(k4,rate2)
poisson2_cumulative
```

```python
#Now we have the sum of probabilities of each occurence,Inorder to find the sum of probability of 5,6,7,8,9,10 cars arriving
#we can subtract the sum of probability of 0 to 4 cars arriving and subtract it from total probability of 1
morethan_4 = 1 - poisson2_cumulative[4]
print("The probability that more than 4 cars will arrive at the station in a given hour = %f" %morethan_4)
```

From the above analysis we can conclude that the charging station that we considered needs an upgradation since it satisfies the criteria. ie poisson2_cumulative > poisson2[4] (0.184737 > 0.16803136)

**Since the data are in different format we need to convert it into a single format 'Funds in Millions'**

```python
def inMillion(x):
    if x[-1]=='K':
        return float(x[1:-1])/1000
    elif x[-1]=='B':
        return float(x[1:-1])*1000
    else:
        return float(x[1:-1])
#This is a function to filter out the different abbreviations and to convert the value to Millions.    
```

```python
copy1_company_data.loc[:,'Funds_in_Millions']= copy1_company_data['Funding'].apply(lambda x:inMillion(x))

```

**Here we took each row of attribute 'Funding' and passed as argument to our function inMillion() using our inline function lambda.**<br>
**We could also do the same with inline function alone. The function takes the string value and check for its abbreviations if it is in 'K' (ie 1000) we divide it with 1000,if it is in 'B' we multiply with 1000 to change the value to Millions and convert it to float.**<br>
**If the abbreviation is 'M' we change it from string to float. In these filtering we also remove the $ symbol from the value.**<br>
**We stored this Filtered values to a new attribute 'Funds_in_Millions'**

**After visual analysis we check for statistical analysis**

**Null Hypothesis (Ho):** There is no significant difference between Funds raised by companies that are still operating and companies that are closed down.<br>
**Alternative Hypothesis (Ha):** There is significant difference between Funds raised by companies that are still operating and companies that are closed down.

**We can either use t-test or z-test for this, but since the sample is significantly large we use z-test**

```python
from statsmodels.stats.weightstats import ztest

s1 = copy1_company_data.loc[copy1_company_data.OperatingState == 'Operating','Funds_in_Millions']
s2 = copy1_company_data.loc[copy1_company_data.OperatingState == 'Closed','Funds_in_Millions']

t_stat,p_value = ztest(s1,s2)

if p_value <= 0.05: #We take the value of alpha as 0.05, ie significance level of 5%
    print(f'We reject the Null Hypothesis since p-value({round(p_value,2)}) is < alpha(0.05) ' )
else:
    print(f'We fail to reject the Null Hypothesis since p-value({round(p_value,2)}) is > alpha(0.05) ' )
```

**<u>Conclusion</u>**
<br>**From above we can conclude that there is not enough evidence to prove there is significant difference between Funds raised by companies that are still operating and companies that are closed down.**

**Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:**

We will use z-test for proportions in this problem

**Null Hypothesis (Ho) :** There is equal proportion of companies that are operating between winners and contestants<br>
**Alternate Hypothesis (Ha) :**  There is significant difference in proportion of companies that are operating between winners and contestants

```python
from statsmodels.stats.proportion import proportions_ztest

test_statistics,pvalue = proportions_ztest([c_operating,w_operating],[t_contestant,t_winner])

if pvalue <= 0.05: #We take the value of alpha as 0.05, ie significance level of 5%
    print(f'We reject the Null Hypothesis since p-value({round(pvalue,2)}) is < alpha(0.05) ' )
else:
    print(f'We fail to reject the Null Hypothesis since p-value({round(pvalue,2)}) is > alpha(0.05) ' )
```

Write and perform your hypothesis along with significance test comparing the funds raised by companies across NY, SF and EU events from
2013 onwards

```python
NY_e = copy1_company_data.loc[d_event_13[d_event_13.apply(lambda x: 'NY' in x)].index, 'Funds_in_Millions']
EU_e = copy1_company_data.loc[d_event_13[d_event_13.apply(lambda x: 'EU' in x or 'London' in x)].index, 'Funds_in_Millions']
SF_e = copy1_company_data.loc[d_event_13[d_event_13.apply(lambda x: 'SF' in x)].index, 'Funds_in_Millions']
```

**Null Hypothesis (Ho):** Funds raised by companies in three different regions are same.<br>
**Alternative Hypothesis (Ha):** Funds raised by companies in three different regions are not same.

**We can plot a distribution plot for the funds raised in three regions for visualizing**

```python
fig2,ax2 = plt.subplots(1,3)
fig2.set_figheight(5)
fig2.set_figwidth(20)

sns.distplot(NY_e,ax=ax2[0],color='Red')

sns.distplot(EU_e,ax=ax2[1],color='Blue')
sns.distplot(SF_e,ax=ax2[2],color='Green')

ax2[0].set_title('Distribution Plot for Funds Raised by Companies in NY')
ax2[1].set_title('Distribution Plot for Funds Raised by Companies in EU')
ax2[2].set_title('Distribution Plot for Funds Raised by Companies in SF')


plt.show()
```

**From above graph we can see the three distribution are quite similar, we can conclude by performing the test**<br>
**There are many test like z-test,t-test etc but since this particular problem contain more than two parameter we can go with ANOVA test**

```python
from scipy.stats import f_oneway

a_stat,a_pvalue = f_oneway(NY_e,SF_e,EU_e)

if a_pvalue <= 0.05:     #In most cases we take significant level of 5%
    print(f'We reject the Null Hypothesis since p-value({round(a_pvalue,3)}) is < alpha(0.05) ' )
else:
    print(f'We fail to reject the Null Hypothesis since p-value({round(a_pvalue,3)}) is > alpha(0.05) ' )
```

**Since there is not enough evidence to reject null hypothesis,we can conclude that funds raised by companies of different regions are significantly higher or lower**
